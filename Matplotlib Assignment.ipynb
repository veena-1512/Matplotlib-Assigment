Q 1 : What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular Python library used for creating data visualizations. It provides a wide range of functions and classes for generating various types of plots and charts. Matplotlib is widely used in the scientific and data analysis communities for visualizing data in a clear and concise manner.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

Line Plot: A line plot displays data points connected by straight line segments. It is commonly used to show the trend or relationship between two variables over a continuous range.

Scatter Plot: A scatter plot represents individual data points as dots on a graph. It is useful for visualizing the distribution and relationship between two variables. Each dot represents the values of two different variables.

Bar Plot: A bar plot, also known as a bar chart, uses rectangular bars to represent data. It is effective for comparing and displaying categorical data. The length of each bar corresponds to the value it represents.

Histogram: A histogram is used to visualize the distribution of a continuous variable. It displays the frequency of values falling into different bins or ranges. The height of each bar represents the number of data points falling within that bin.

Pie Chart: A pie chart is a circular graph divided into slices, where each slice represents a different category. It is commonly used to display the proportion or percentage of each category in a dataset.

Q 2. A scatter plot is a type of data visualization that displays the relationship between two variables. It is particularly useful for identifying patterns, trends, or correlations between the variables. In a scatter plot, each data point is represented by a marker on a Cartesian coordinate system, with one variable represented on the x-axis and the other variable represented on the y-axis.

To generate data for the variables x and y using the given code:

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

import matplotlib.pyplot as plt

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```

Q 4. A bar plot, also known as a bar chart or bar graph, is a visualization that represents categorical data using rectangular bars. The length of each bar corresponds to the value of the data it represents. Bar plots are commonly used to compare and display the relative values or frequencies of different categories.

Bar plots are useful for several reasons:

Comparison: Bar plots make it easy to compare the values of different categories visually. The length or height of the bars provides a clear representation of the magnitude of each category's value.

Categorical data: Bar plots are especially useful when dealing with categorical or discrete data. They are effective for showing the distribution or frequencies of different categories.

Visualization of relationships: Bar plots can also be used to show relationships between categories or subcategories within a larger group. For example, multiple bars can be used to compare subcategories within each main category.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")
plt.title("Profit of Companies")
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 5))  # Set the figure size (optional)
plt.barh(company, profit)
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")
plt.title("Profit of Companies")
plt.show()
```

Q 5.A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It summarizes the key statistical measures of the data, including the minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum. The plot consists of a box that spans the interquartile range (IQR) with a line inside representing the median. Whiskers extend from the box to the minimum and maximum values within a specified range, typically 1.5 times the IQR. Data points that fall outside this range are considered outliers and are represented as individual points.

Box plots are used for several purposes:

Summary of distribution: Box plots provide a concise visual summary of the distribution of a dataset. They show the spread, central tendency, and presence of outliers in the data.

Comparison: Box plots are useful for comparing the distributions of multiple datasets or groups. They make it easy to identify differences in the medians, quartiles, and ranges between different groups.

Outlier detection: Box plots help identify potential outliers in the dataset. Outliers are data points that significantly deviate from the rest of the data and may indicate anomalies or errors.

An example of how to create a box plot using the given data:

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig, ax = plt.subplots()

ax.boxplot(data)

ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')

# Set the title
ax.set_title('Box Plot')

# Display the plot
plt.show()
```
